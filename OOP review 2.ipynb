```python
class Employee:
    
    raise_amt = 1.04
    
    def __init__(self, first, last, pay): #attributes of our class
        self.first = first #done automatically now
        self.last = last
        self.pay = pay
        self.email = first + '.' + last + '@company.com'
        
    #method call fullname, pass arguement(self)
    def fullname(self):
        return "{} {}".format(self.first, self.last)
    
    def apply_raise(self):
        self.pay = int(self.pay * self.raise_amt)
    
    #inherit everything from employee
class Developer(Employee):
    raise_amt= 1.10
    
    def __init__(self, first, last, pay, prog_lang):
        super().__init__(first, last, pay) #from employee
        self.prog_lang = prog_lang
        
        
class Manager(Employee):
    
    def __init__(self, first, last, pay, employees=None):
        super().__init__(first, last, pay) #from employee
        if employees is None:
            self.employees = []
        else:
            self.employees = employees
    def add_emp(self, emp):
        if emp not in self.employees:
            self.employees.append(emp)
            
    def remove_emp(self,emp):
        if emp in self.employees:
            self.employees.remove(emp)
            
    def print_emps(self):
        for emp in self.employees:
            print('-->', emp.fullname())
    
        
dev_1 = Developer('Hong','Gao',5000, 'Python') 
dev_2 = Developer('Lo','Man',5000, 'Java')

mgr_1 = Manager('Sue', 'Smith', 9000, [dev_1])

print(mgr_1.email)

mgr_1.add_emp(dev_2)

mgr_1.print_emps()

#print(help(Developer)) #first it didnt find in developer, then look to employee class
#print(dev_1.email)
#print(dev_2.email)

#print(dev_1.email)
#print(dev_1.prog_lang)

#print(dev_1.pay)
#dev_1.apply_raise()
#print(dev_1.pay)

#tells us if an instance is of a class
isinstance(mgr_1, Developer)

issubclass(Developer, Employee)


```
